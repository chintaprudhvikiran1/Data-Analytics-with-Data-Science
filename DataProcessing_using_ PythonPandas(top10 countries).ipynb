```python
# Checking the Region column to see if it contians 'Asia' or 'Americas'
# isin function
countries_df['Region'].isin(['Asia','Americas'])
```

```python
# Returning only rows where Region='Asia' or Region='Americas'
countries_df[countries_df['Region'].isin(['Asia','Americas'])]
```

```python
# Returning only rows where Region is NOT'Asia' and Region is NOT'Americas'
countries_df[~countries_df['Region'].isin(['Asia','Americas'])]
```

```python
tfl_df = pd.read_csv('cycles_on_rent.csv')
```

```python
tfl_df.head()
```

```python
# Clean data. remove last column
tfl_df.drop(columns='Unnamed: 2', inplace=True)
```

```python
tfl_df.head()
```

```python
tfl_df.shape
```

```python
tfl_df.dtypes
```

```python
# Changing date date in Day column using datetime
tfl_df['Day']=tfl_df['Day'].astype('datetime64[ns]')
```

```python
tfl_df.head()
```

```python
tfl_df.dtypes
```

```python
# formatting output to show only month and year columns as string
tfl_df['Day']= pd.to_datetime(tfl_df['Day'],format='%d/%m/%Y')
```

```python
tfl_df.dtypes
```

```python
tfl_df.head()
```

```python
# sorting values by the 'Number of Bicycle Hires' in descending order
tfl_df.sort_values(by='Number of Bicycle Hires', ascending=False, inplace=True)
```

```python
tfl_df.head()
```

```python
import datetime as dt
```

```python
# strftime is a method in the datetime module that converts a datetime object to string and outputs in a specified format
tfl_df['Month / Year'] = tfl_df['Day'].dt.strftime('%b-%y')
```

```python
tfl_df
```

```python
# the transpose() method transposes the rows and columns, the index row becomes the column headers
tfl_df.transpose()
```

```python
# setting the index to the 'Month / Year' column
tfl_df.set_index(keys='Month / Year', inplace=True)
```

```python
tfl_df.transpose()
```

##### explode function

```python
column_index = ['A','B']
data_list = [[[1,2], 'nested list'],[2,'not a nested list'],[[1,2],'nested list 2']]
```

```python
df=pd.DataFrame(data=data_list, columns=column_index)
df
```
