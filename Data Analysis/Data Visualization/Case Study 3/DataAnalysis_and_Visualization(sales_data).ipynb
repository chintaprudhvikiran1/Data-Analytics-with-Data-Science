### Practice Project

**Project Title:** Analysis of Sales Data for a Retail Store

**Objective:** Learn how to clean, analyze, and visualize sales data using Python, focusing on deriving insights such as top-selling products, sales trends over time, and geographical sales performance

##### 1. Importing and Understanding Data

```python
import pandas as pd
# Load the dataset
data = pd.read_csv('sales_data.csv')

# Display the first few rows
print(data.head())

# Check data information
print(data.info())
```

```python
data.head()
```

```python
data.describe()
```

##### 2. Cleaning the Data

- Handle missing values.
- Convert data types (e.g., parse Date column as datetime).
- Remove duplicate rows.

```python
# Check for missing values
print(data.isnull().sum())

# Drop rows with missing data
data.dropna(inplace=True)

# Convert 'Date' to datetime
# data['Date'] = pd.to_datetime(data['Date'])

# Drop duplicates
data.drop_duplicates(inplace=True)
```

##### 3. Exploratory Data Analysis (EDA)

- Total Sales by Category:

```python
sales_by_category = data.groupby('Category')['Total Price'].sum()
print(sales_by_category)
```

- Top-Selling Products:

```python
top_products = data.groupby('Product')['Quantity'].sum().sort_values(ascending=False).head(5)
print(top_products)
```

- Sales Trend Over Time:

```python
sales_trend = data.groupby('Date')['Total Price'].sum()
print(sales_trend)
```

#### 4. Data Visualization

- Bar Chart for Total Sales by Category:

```python
import matplotlib.pyplot as plt

sales_by_category.plot(kind='bar', color='skyblue')
plt.title('Total Sales by Category')
plt.ylabel('Total Sales ($)')
plt.show()
```

- Line Plot for Sales Trend Over Time:

```python
sales_trend.plot(kind='line', marker='o', color='green')
plt.title('Sales Trend Over Time')
plt.xlabel('Date')
plt.ylabel('Total Sales ($)')
plt.grid(True)
plt.show()
```

- Top Products as a Horizontal Bar Chart:

```python
top_products.plot(kind='barh', color='orange')
plt.title('Top-Selling Products')
plt.xlabel('Quantity Sold')
plt.show()
```
