```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
def demonstrate_central_limit_theorem(sample_size=30, num_experiments=1000):
    """
    Demonstrate the Central Limit Theorem (CLT)
    
    Parameters:
    -----------
    sample_size : int
        Number of samples in each experiment
    num_experiments : int
        Number of times the sampling is repeated
    """
    # Set random seed for reproducibility
    np.random.seed(42)
    
    # Three different underlying distributions
    distributions = {
        'Uniform Distribution': np.random.uniform(0, 10, (num_experiments, sample_size)),
        'Exponential Distribution': np.random.exponential(2, (num_experiments, sample_size)),
        'Skewed Distribution': np.random.chisquare(3, (num_experiments, sample_size))
    }
    
    # Create a figure with subplots
    fig, axes = plt.subplots(len(distributions), 2, figsize=(15, 15))
    fig.suptitle('Central Limit Theorem Demonstration', fontsize=16)
    
    # Iterate through different distributions
    for i, (dist_name, dist_data) in enumerate(distributions.items()):
        # Calculate sample means
        sample_means = np.mean(dist_data, axis=1)
        
        # Original Distribution
        axes[i, 0].hist(dist_data[0], bins=30, density=True, alpha=0.7, color='blue')
        axes[i, 0].set_title(f'Original {dist_name}')
        axes[i, 0].set_xlabel('Value')
        axes[i, 0].set_ylabel('Frequency')
        
        # Distribution of Sample Means
        axes[i, 1].hist(sample_means, bins=30, density=True, alpha=0.7, color='red')
        axes[i, 1].set_title(f'Distribution of Sample Means\n(n={sample_size})')
        axes[i, 1].set_xlabel('Sample Mean')
        axes[i, 1].set_ylabel('Frequency')
        
        # Overlay normal distribution on sample means
        x = np.linspace(sample_means.min(), sample_means.max(), 100)
        pdf = stats.norm.pdf(x, np.mean(sample_means), np.std(sample_means))
        axes[i, 1].plot(x, pdf, 'r-', lw=2, label='Normal Distribution')
        axes[i, 1].legend()
    
    plt.tight_layout()
    plt.show()
    
    # Print statistical insights
    print("\nCentral Limit Theorem Insights:")
    for dist_name, dist_data in distributions.items():
        # Calculate statistics of original distribution
        orig_mean = np.mean(dist_data)
        orig_std = np.std(dist_data)
        
        # Calculate statistics of sample means
        sample_means = np.mean(dist_data, axis=1)
        means_mean = np.mean(sample_means)
        means_std = np.std(sample_means)
        
        print(f"\n{dist_name}:")
        print(f"Original Distribution:")
        print(f"  Mean: {orig_mean:.4f}")
        print(f"  Standard Deviation: {orig_std:.4f}")
        print(f"Distribution of Sample Means:")
        print(f"  Mean of Means: {means_mean:.4f}")
        print(f"  Standard Deviation of Means: {means_std:.4f}")
```

```python
demonstrate_central_limit_theorem(sample_size=30, num_experiments=1000)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Step 1: Create a non-normal (right-skewed) population
population = np.random.exponential(scale=2, size=10000)  # Exponential distribution (skewed)

# Plot the original population
plt.figure(figsize=(10, 5))
sns.histplot(population, bins=50, kde=True, color='#343A40')
plt.title("Original Population Distribution (Skewed)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()

# Step 2: Demonstrate CLT by taking multiple samples and calculating means
sample_size = 30  # Size of each sample
num_samples = 1000  # Number of samples

sample_means = []  # Store means of all samples
for i in range(num_samples):
    sample = np.random.choice(population, size=sample_size)  # Take a random sample
    sample_mean = np.mean(sample)  # Calculate mean of the sample
    sample_means.append(sample_mean)  # Append to the list

# Step 3: Plot the distribution of sample means
plt.figure(figsize=(10, 5))
sns.histplot(sample_means, bins=50, kde=True, color='#20C997')
plt.title("Distribution of Sample Means (CLT in Action)")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.show()

# Step 4: Compare the original population with the sample means
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# Plot original population
sns.histplot(population, bins=50, kde=True, color='#343A40', ax=axes[0])
axes[0].set_title("Original Skewed Population")
axes[0].set_xlabel("Value")
axes[0].set_ylabel("Frequency")

# Plot sample means
sns.histplot(sample_means, bins=50, kde=True, color='#20C997', ax=axes[1])
axes[1].set_title("Sample Means (Approximately Normal)")
axes[1].set_xlabel("Sample Mean")
axes[1].set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```
