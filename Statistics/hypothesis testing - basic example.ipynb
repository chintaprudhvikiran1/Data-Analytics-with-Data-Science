```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Given data
p_0 = 0.30  # Null hypothesis proportion (your email open rate)
n = 100  # Sample size of competitor's emails
p_c = 38 / 100  # Competitor's observed open rate (38 opened out of 100 emails)
alpha = 0.05  # Significance level

# Step 1: Calculate the test statistic
z_stat = (p_c - p_0) / np.sqrt((p_0 * (1 - p_0)) / n)

# Step 2: Calculate the critical value (z-critical for one-tailed test)
z_critical = norm.ppf(1 - alpha)

# Step 3: Calculate the p-value (one-tailed test)
p_value = 1 - norm.cdf(z_stat)

# Print results and insights
print("Hypothesis Test Results:")
print(f"Observed Proportion (Competitor's open rate): {p_c:.2f}")
print(f"Null Hypothesis Proportion (Your open rate): {p_0:.2f}")
print(f"Sample Size: {n}")
print(f"Z-statistic: {z_stat:.4f}")
print(f"Z-critical (at alpha = {alpha}): {z_critical:.4f}")
print(f"P-value: {p_value:.4f}")

# Making Decision
if p_value <= alpha:
    print("Conclusion: Reject the null hypothesis. The competitor's open rate is significantly higher than yours.")
else:
    print("Conclusion: Fail to reject the null hypothesis. No evidence that the competitor's open rate is higher.")

# Visualization: Standard normal distribution and rejection region
z_values = np.linspace(-4, 4, 1000)  # Z-values for the normal curve
pdf = norm.pdf(z_values)  # Probability density function of the standard normal distribution

plt.figure(figsize=(10, 6))
plt.plot(z_values, pdf, label="Standard Normal Distribution", color='blue')

# Shade the rejection region
plt.fill_between(z_values, 0, pdf, where=(z_values >= z_critical), color='red', alpha=0.4, label="Rejection Region")

# Mark the z-statistic
plt.axvline(z_stat, color='green', linestyle='--', label=f"Z-statistic = {z_stat:.2f}")

# Add critical value line and annotations
plt.axvline(z_critical, color='red', linestyle='--', label=f"Z-critical = {z_critical:.2f}")
plt.text(z_stat + 0.1, 0.1, "Z-statistic", color="green")
plt.text(z_critical + 0.1, 0.02, "Z-critical", color="red")

# Add labels and legend
plt.title("Hypothesis Test Visualization: One-Tailed Test", fontsize=14)
plt.xlabel("Z-value", fontsize=12)
plt.ylabel("Probability Density", fontsize=12)
plt.legend(fontsize=12)
plt.grid(alpha=0.3)
plt.show()
```
